```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,31)
x
```

```python
y=random.randint(1,100,30)
y
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot")
```
